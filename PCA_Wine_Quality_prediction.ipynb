```python
#Defining feature and target

x=data.drop(columns=['quality'])
y=data['quality']
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
ss=StandardScaler()
scaled_data=ss.fit_transform(x)
scaled_data=pd.DataFrame(scaled_data,columns=x.columns)
scaled_data.head()
```

```python
#applying PCA

pca=PCA().fit(scaled_data)
plt.rcParams['figure.figsize']=(12,6)
fig,ax=plt.subplots()

xi=np.arange(1,12,step=1)
yi=np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)

plt.plot(xi,yi,linestyle='--',color='b',marker='o')

plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.xticks(np.arange(0,12,step=1))
plt.title('The number of components needed to explain the variance')

plt.axhline(y=0.97,color='r',linestyle='-')
plt.text(0.1,0.85,'97% threshold',color='red',fontsize=16)
plt.grid(axis='x')
plt.show()

```

```python
#perform PCA

pca=PCA(n_components=9)
pca.fit(scaled_data)

x_pca=pca.transform(scaled_data)
x_pca
```

```python
x_pca[0]
```

```python
pca.components_.shape
```

```python
# converting x_pca to dataframe

df_pca=pd.DataFrame(x_pca,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9'])
df_pca.head()
```

```python
df_pca.shape
```

```python
df_pca['target']=y
df_pca.head(3)
```

```python
#splitt

x1=df_pca.drop(columns=['target'])
y1=df_pca['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
#Considering the target as a continuous num, so this will be done as a regression problem
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
df_1=pd.DataFrame({'y_test':y_test,"y_pred":y_pred})
sns.lineplot(data=df_1)
plt.show()
```
