```python
outlier_columns=['spirit_servings', 'wine_servings']
#calculation the quartile and IQR

q1=data[outlier_columns].quantile(0.25)
q3=data[outlier_columns].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
#treating the outliers

outlier_lower=data[outlier_columns] < (q1-1.5*IQR)
outlier_upper=data[outlier_columns] > (q3+1.5*IQR)

#outlier clip
data[outlier_columns]=data[outlier_columns].clip(lower=q1-1.5*IQR, upper=q3+1.5*IQR,axis = 1)
```

```python
numerical_columnss=['spirit_servings', 'wine_servings']
plt.figure(figsize=(15,10))

sns.boxplot(data=data[numerical_columnss])
plt.title("Recheck for outliers")
plt.ylabel("value")
plt.xlabel("Numerical columns")
plt.xticks(rotation=45,fontsize=20)
plt.show()
```
